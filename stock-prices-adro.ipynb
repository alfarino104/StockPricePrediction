**Splitting the training data**

```python
df_n = df.copy()
y_n = y.copy()
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid= train_test_split(df_n, y_n, test_size= 0.2)

print(f'Train dataset: {X_train.shape}, Train target dataset: {y_train.shape}')
print(f'Valid dataset: {X_valid.shape}, Valid target dataset: {y_valid.shape}')
```

**Normalize the dataset**

```python
#import the library
from sklearn.preprocessing import StandardScaler

#Fit the standardization scaler onto the training data
stan= StandardScaler().fit(X_train)

#Transform the training data
X_train_stand= stan.transform(X_train)
X_valid_stand= stan.transform(X_valid)

#Convert the transformed data into pandas dataframe 
X_train_stand_df= pd.DataFrame(X_train_stand)
X_valid_stand_df= pd.DataFrame(X_valid_stand)

#Assigning original feature names for ease of read
X_train_stand_df.columns= X_train.columns
X_valid_stand_df.colums= X_valid.columns

X_train_stand_df.describe()
```

```python
X_train_stand_df= X_train_stand_df.fillna(0)
```

```python
X_train_stand_df
```

**Training Process**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model= LinearRegression()
model.fit(X_train_stand_df, y_train)
```

```python
X_valid_stand_df= X_valid_stand_df.fillna(0)
```

```python
y_pred_validation= model.predict(X_valid_stand_df)

mse_validation= mean_squared_error(y_valid, y_pred_validation)
print(f'MSE on Validation dataset: {mse_validation}')
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

mae_validation= mean_absolute_error(y_valid, y_pred_validation)
print(f'MAE on Validation dataset: {mae_validation}')

rmse_validation= np.sqrt(mse_validation)
print(f'RMSE on Validation dataset: {rmse_validation}')

r2_score_validation= r2_score(y_valid, y_pred_validation)
print(f'R2 Score on Validation dataset: {r2_score_validation}')
```
